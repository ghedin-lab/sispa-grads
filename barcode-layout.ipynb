## Barcodes Only

We can start by looking at only the barcode alignments within the first 50 bp. This range will give us both better resolution as well as the ability to detect whether or not (and presumably to what degree) there has been any cleavage.

```r
bc50 = bc[which(bc$start <= 50),]
hist(x=bc50$start[which(bc50$bc != "sa")], nclass=length(unique(bc50$start)))
```

Only 900,000 at position 5 which is where we would expect the barcodes to be.

## Standard Amp Only

May as well plot the std amp and see if that clears anything up.

```r
hist(x=bc50$start[which(bc50$bc == "sa")], nclass=length(unique(bc50$start)))
```

This is reassuring, though the yield seems terrible. Kristen used bwa fastmap with the entire std amp sequence which may have skewed the results, but given this isn't much larger than what we've seen in the barcodes only, I'm not sure if it'll get much better.

However, based on the fastqc reports, there definitely seems to be some consistency with the first 5 bases of each read: http://core-fastqc.bio.nyu.edu/HYGNVBGXX/merged/

We will need to map the 5bp terminus of the standard amp to the reads to see if there is a greater yield. But first, we can check to see whether these are complete alignments based on the input data.

```r
head(bc[which(bc$bc == "sa"),])
```

```r
hist(x=bc$end - bc$start, nclass=length(unique(bc$end - bc$start)))
```

This is reassuring - the majority of alignment lengths are 7bp. Now to look at everything greater than 11bp to get a sense of chaining.

```r
bc$aln_len = bc$end - bc$start
temp=bc$aln_len[which(bc$aln_len >= 11)]
hist(x=temp, nclass=length(unique(temp)),xlim=c(11,30))
```

There are ~ 500,000 full SA alignments and 3,000,000 11 bp alignments. We should see if they're aligning in any consistent fashion.

```r
hist(bc$start[which(bc$aln_len == 11)], nclass=length(unique(bc$start[which(bc$aln_len == 11)])))
```

So nothing seems very consistent which is fairly reassuring, though the two tallest peaks are roughly the same distance apart if there was chaining, but these mapping results seem wonky so I'm running it again with bwa mem.
